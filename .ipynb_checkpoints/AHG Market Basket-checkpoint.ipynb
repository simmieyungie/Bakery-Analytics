```python
#modules
#modules
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import pyodbc
```

The adventureWorks2012 database was used  for this analysis. Data was queried directly with python from the SQL Server the database was hosted.

```python
#import data from sql db
conn = pyodbc.connect("Driver={SQL Server};\
                     Server=SIMIPV;\
                     PWD=Password;Database=AdventureWorks2012;")

#query
df = (pd.read_sql_query("SELECT \
                        SOH.[OnlineOrderFlag], \
                        SOH.[CustomerID], \
                        PP.[Name], \
                        PSC.[Name] as [ProductSubName], \
                        PC.[Name] as [ProductCatName], \
                        SOD.[OrderQty], \
                        SOH.[OrderDate] \
                        FROM [Sales].[SalesOrderHeader] AS [SOH] \
                        INNER JOIN [Sales].[SalesOrderDetail] AS [SOD] \
                        ON [SOH].[SalesOrderID] = SOD.[SalesOrderID] \
                        INNER JOIN [Production].[Product] AS [PP] \
                        ON [SOD].[ProductID] = [PP].ProductID \
                        INNER JOIN [Production].[ProductSubcategory] AS [PSC] \
                        ON [PP].ProductSubcategoryID = PSC.ProductSubcategoryID \
                        INNER JOIN [Production].[ProductCategory] AS [PC] \
                        ON [PSC].ProductCategoryID = [PC].ProductCategoryID \
                        ORDER BY [CustomerID]", conn) )

#write data to csv
df.to_csv("transactionsAHW.csv", index = False)
#preview data
df.head()
```

```python
#read in data
df = pd.read_csv("transactionsAHW.csv")

#preview
df.head()
```

```python
#Filter for online transactions only
df = df[df["OnlineOrderFlag"] == True]
```

### Data Cleaning and Agg

```python
#Change CustomerID to string
df["CustomerID"] = df.CustomerID.map(str)
```

```python
#summerise data
basket = (df.groupby(["CustomerID", "ProductSubName"])["OrderQty"]
.sum().unstack().reset_index().fillna(0).set_index("CustomerID"))


#preview
basket.head()
```

```python
basket.columns
```

```python
#remap values to 0 and 1
def encode(x):
    if x<= 0:
        return 0
    else:
        return 1
    
#use function
basket_sets = basket.applymap(encode)
```

#### Market Basket Analysis

### Support:
    - In Market Basket analysis, support represents the popularity of an item in all transactions. Support of an item is calculated as the number of transactions involving that item and the total number of transactions. 
    Support of an item = *(Number of transactions involving that item)/
                            (Total number of transactions)*

```python
#frequent itemsets
frequent_items = apriori(basket_sets, min_support = 0.07, use_colnames = True)

frequent_items.sort_values(by = "support", ascending = False).head()
```

```python
#Generate the rules
rules = association_rules(frequent_items, metric = "lift", min_threshold = 1)

rules.head(10) #preview
```

```python
rules.sort_values(by = "lift",ascending = False)
```

```python
#filter for lift greater than 1 and confidence >= 0.3
rules[(rules['lift'] > 1) &
        (rules["confidence"] >= 0.3)]
```
